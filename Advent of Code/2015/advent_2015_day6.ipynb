```python
## Tiny little helper function that as bigger before, but i cut it down.
def get_cordinate_pair(cordinate_text: str):
    cordinates = [int(x) for x in cordinate_text.split(",")]
    return cordinates
```

```python
## Day 6 part 1 - Used numpy to make it simpler
def turning_on_the_lights(led_matrix, line_input : str):
    #Split the line so i can use it
    order_list = line_input.split()
    ## The reason i use these start and end list containers instead of some smarter looking code is that is easier for me to follow
    start = []
    end = []

    # Here we check the if the instructions use "toggle".
    # If True, then the turn on/off string split has one more word so the index is moved by 1
    if order_list[0] == "toggle":
        start = get_cordinate_pair(order_list[1])
        end = get_cordinate_pair(order_list[3])
        ## Found out numpy has a clever logic toggle that can be used with booleans
        led_matrix[start[1]:end[1]+1,start[0]:end[0]+1] = np.logical_not(led_matrix[start[1]:end[1]+1,start[0]:end[0]+1])
    else:
        start = get_cordinate_pair(order_list[2])
        end = get_cordinate_pair(order_list[4])
        ## Setting the instructed command
        if order_list[1] == "off":
            led_matrix[start[1]:end[1]+1,start[0]:end[0]+1] = False
        elif order_list[1] == "on":
            led_matrix[start[1]:end[1]+1,start[0]:end[0]+1] = True


## Day 6 part 2 - I could reuse every thing and only change the operations on the sub matrices, thats why we make functions.
## As the matrix will not not be a boolean matrix, but a 1000x1000 int matrix intialised as 0 
def turning_on_the_lights_brightness(led_matrix, line_input : str):
    order_list = line_input.split()
    start = []
    end = []
    # Here we check the toggle.
    if order_list[0] == "toggle":
        start = get_cordinate_pair(order_list[1])
        end = get_cordinate_pair(order_list[3])
        led_matrix[start[1]:end[1]+1,start[0]:end[0]+1] += 2
    else:
        start = get_cordinate_pair(order_list[2])
        end = get_cordinate_pair(order_list[4])
        if order_list[1] == "off":
            led_matrix[start[1]:end[1]+1,start[0]:end[0]+1] -= 1 ## Should maybe set it to += -1 as it seem more right, but that might only be the tylanol speaking through me
            ## If the value gets below Zero then it will be set to Zero. 
            #led_matrix[led_matrix < 0] = 0 ## This works 
            np.clip(led_matrix, 0, None, out=led_matrix) ## Read a bit and this should be faster....
        elif order_list[1] == "on":
            led_matrix[start[1]:end[1]+1,start[0]:end[0]+1] += 1
```

```python
## Day 6 Part 1.
## Creating the LED matrix with a million, i set the type to Bool so it would be a 
## True/False matrix to allow for some smart operations
led_matrix = np.zeros((1000, 1000),dtype=bool) 
with open(input_path,'r') as text:
    for line in text:
        turning_on_the_lights(led_matrix,line)

## Going to be a bit clever and only count non zeros in the matrix, thank you Bool operations
print(f"The number of on LEDS is {np.count_nonzero(led_matrix)}")
        
```

```python
## Day 6 Part 2.
## The only real change will be the type of matrix
led_matrix_brightness = np.zeros((1000, 1000),dtype=int) ## Techinally no type is needed, but i like it this way and it is a good habbit
with open(input_path,'r') as text:
    for line in text:
        turning_on_the_lights_brightness(led_matrix_brightness,line) ## Rember run the right function

## We can use sum to find out the culumanative brightness.
print(f"The Brightness of the LEDS is {np.sum(led_matrix_brightness)}")
```
