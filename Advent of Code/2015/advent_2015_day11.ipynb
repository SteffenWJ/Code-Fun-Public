The Rules:  
Password most be 8 lowercase leters long  
Passwords must include one increasing straight of at least three letters, like abc, bcd, cde, and so on, up to xyz. They cannot skip letters; abd doesn't count.  
Passwords may not contain the letters i, o, or l, as these letters can be mistaken for other characters and are therefore confusing.  
Passwords must contain at least two different, non-overlapping pairs of letters, like aa, bb, or zz.

So go from right most and keep going until we hit Z and then change it. Durring each increment we do the cascade failture, if the "new" password failes we keep incerementing until we got a working one

```python
## One of the test to see if the password lives up to the rules
def test_increasing_straight(current_password : list) -> bool:
    for i in range(len(current_password)-2):
        if current_password[i] == 'z' or current_password[i] == 'y':
            continue ## Made a msitake earlier by breaking out of the loop isntead of skipping
        secound_letter = chr(ord(current_password[i])+1)
        third_letter = chr(ord(current_password[i])+2)
        if current_password[i+1] == secound_letter and current_password[i+2] == third_letter:
            return True
    return False


## Testing if there are no forbiden letters
def no_forbiden_letters(current_password : list ) -> bool:
    forbiden_letters = ['i','o','l']
    for i in range(len(current_password)):
        if current_password[i] in forbiden_letters: ## Easy enough way to do a quick look
            return False
    return True

## Testing if there a atleast two non overlapping pairs
def contains_two_different_overlaping(current_password : list) -> bool:
    i = 0
    number_of_overlaping = 0
    found_letters = []
    while i < len(current_password)-1: ## Firsty foor loop so i can move the iterator agian
        if current_password[i] == current_password[i+1] and current_password[i] and current_password[i] not in found_letters:
            found_letters.append(current_password[i])
            number_of_overlaping += 1
            i += 2 ## If we find a pair we move the iterator forward to skip
        else:
            i += 1
    if number_of_overlaping >= 2:
        return True
    else:
        return False
```

```python
## A wrapper function to check if the new password is good enough.
def run_the_test(current_password : list) -> bool:
    ## This is not needed, we could do a direct check, but i like it this way
    test_1 = test_increasing_straight(current_password)
    test_2 = no_forbiden_letters(current_password)
    test_3 = contains_two_different_overlaping(current_password)    
    if test_1 and test_2 and test_3:
        return True
    else:
        return False
```

```python
def incrementing_the_password(password : list):
    for i in range(len(password)-1,-1, -1): ## I hate reverse for loops, remeber not 0, but -1 becaus it is the one before.
        if password[i] == 'z':
            password[i] = 'a'
        else:
            password[i] = chr(ord(password[i])+1)
            break
    return password
```

```python
## Created this function to run my code a bit more compact.
def get_new_password(password_input : str) -> str:
    password_list = list(password_input) ## stupid strings being immutable
    old_password = list(password_input) ## Added this for part 2, so i can comapre a list to a list
    passed_test = False
    while True:
        passed_test = run_the_test(password_list)
        if passed_test and old_password != password_list: ## For part two i added a quick test to see if the the new password was different form the old
            suggested_pass = "".join(password_list)
            return suggested_pass
        password_list = incrementing_the_password(password_list)
```
