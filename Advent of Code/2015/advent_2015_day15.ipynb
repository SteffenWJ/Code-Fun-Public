```python
input_path = ""
```

```python
## I feel like i make a lot of dicts
## I am also adding colories in this even if it is not needed to make it simpler for part to, i assume i will need it.
def create_ingredient_dict(ingredient_name: str, capacity :int , durability : int, flavor : int, texture: int, calories : int,  the_dict: dict):
    the_dict[ingredient_name] = {"capacity" : capacity, "durability" : durability, "flavor" : flavor, "texture" : texture, "calories" : calories}
```

```python
## You know it, you love it, the same kind of stuff as before.
ingredients_dict = {}
with open(input_path, 'r') as text:
    for line in text:
        line_split = line.split()
        #print(line_split)
        ing_name = line_split[0].strip(':')
        ing_capacity = int(line_split[2].strip(','))
        ing_dura = int(line_split[4].strip(','))
        ing_flavor = int(line_split[6].strip(','))
        ing_texture = int(line_split[8].strip(','))
        ing_calories = int(line_split[10].strip(','))
        create_ingredient_dict(ing_name,ing_capacity,ing_dura,ing_flavor,ing_texture,ing_calories,ingredients_dict)
```

```python
## SOOO this is ugly. It is a brute force way to do it that is in no way optimal in terms of time complexity, 

def brute_force_cookies(ingredients_dict):
    capacity_sum = 0
    durrabilty_sum = 0
    flavor_sum = 0
    texture_sum = 0
    sum_value = 0
    best_value = 0
    best_mix = ""

    for idx_A in range(0,101): ## Frosting
        for idx_B in range(0,101): ## Candy
            for idx_C in range(0,101): ## Butter
                for idx_D in range(0,101): ## Sugar
                    if idx_A+idx_B+idx_C+idx_D == 100:
                        capacity_sum = (ingredients_dict["Frosting"]["capacity"]*idx_A+
                                    ingredients_dict["Candy"]["capacity"]*idx_B+
                                    ingredients_dict["Butterscotch"]["capacity"]*idx_C+
                                    ingredients_dict["Sugar"]["capacity"]*idx_D)
                        capacity_sum   = max(0, capacity_sum)  ## One of the issues i ran into is that i forgot to add the clamping in cases where the sum was negative as stated in the instructions.
                        
                        durrabilty_sum = (ingredients_dict["Frosting"]["durability"]*idx_A+
                                    ingredients_dict["Candy"]["durability"]*idx_B+
                                    ingredients_dict["Butterscotch"]["durability"]*idx_C+
                                    ingredients_dict["Sugar"]["durability"]*idx_D)
                        durrabilty_sum = max(0, durrabilty_sum)
                        
                        flavor_sum = (ingredients_dict["Frosting"]["flavor"]*idx_A+
                                    ingredients_dict["Candy"]["flavor"]*idx_B+
                                    ingredients_dict["Butterscotch"]["flavor"]*idx_C+
                                    ingredients_dict["Sugar"]["flavor"]*idx_D)
                        flavor_sum     = max(0, flavor_sum)
                        
                        texture_sum = (ingredients_dict["Frosting"]["texture"]*idx_A+
                                    ingredients_dict["Candy"]["texture"]*idx_B+
                                    ingredients_dict["Butterscotch"]["texture"]*idx_C+
                                    ingredients_dict["Sugar"]["texture"]*idx_D)
                        texture_sum    = max(0, texture_sum)
                        
                        ## Added the best mix for fun, it is not needed
                        sum_value = capacity_sum*durrabilty_sum*flavor_sum*texture_sum
                        if sum_value > best_value:
                            best_value = sum_value
                            best_mix = "Frosting: "+str(idx_A)+" Candy: "+str(idx_B)+" Butter: "+str(idx_C)+" Sugar: "+str(idx_D)

    return best_mix , best_value
```

```python
## Day 15 part 1
best_mix, flavor_sum = brute_force_cookies(ingredients_dict)
print(f"The best mix is {best_mix} with total off {flavor_sum}")
```
