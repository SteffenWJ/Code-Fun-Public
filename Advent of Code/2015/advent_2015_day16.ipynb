Advent of Code 2015 day 16 Part 1

```python
## Part day 16 part 1
## Created a dict for the target so i can pass it to the function checking the line later.
target_sue_dict = {"target":
                   {"children:": 3,"cats:": 7,"samoyeds:": 2,"pomeranians:": 3,"akitas:": 0,"vizslas:": 0,
                    "goldfish:": 5,"trees:": 3,"cars:": 2,"perfumes:": 1}
                    }
```

```python
input_path = ""
```

```python
## Simple enough function, i check if there is a failure and if there is a escape early, and if not that means we foudn the right Sue
def dose_the_sue_match(text_line : str, target_dict : dict):
    entry_split = text_line.split()
    for i in range(2, len(entry_split),2): ## Start index 2, and skip 2
        if target_dict["target"][entry_split[i]] != int(entry_split[i+1].strip(',')):
            return False
    return True
```

```python
## Advent of Code 2015 day 16 part 1
with open(input_path, 'r') as text:
    found_sue = False
    for line in text:
        found_sue = dose_the_sue_match(line, target_sue_dict)
        if found_sue:
            temp_split = line.split()
            print(f"Sue : {temp_split[1].strip(':')} was the right one")
            break
```

Advent of code 2015 Day 16 Part 2

Easiest way is to keep it as it is, but have 4 special cases, should handle it

```python
def dose_the_sue_match_special_cases(text_line : str, target_dict : dict):
    entry_split = text_line.split()
    for i in range(2, len(entry_split),2): ## Start index 2, and skip 2
        if entry_split[i] in ("cats:", "trees:"):  ## Added the first special case check
            if target_dict["target"][entry_split[i]] >= int(entry_split[i+1].strip(',')):
                return False
        elif entry_split[i] in ("pomeranians:", "goldfish:"): ## Added the secound special case check
            if target_dict["target"][entry_split[i]] <= int(entry_split[i+1].strip(',')):
                return False
        else:
            if target_dict["target"][entry_split[i]] != int(entry_split[i+1].strip(',')):
                return False
    return True
```

```python
## Advent of Code 2015 day 16 part 1
with open(input_path, 'r') as text:
    found_sue = False
    for line in text:
        found_sue = dose_the_sue_match_special_cases(line, target_sue_dict)
        if found_sue:
            temp_split = line.split()
            print(f"Sue : {temp_split[1].strip(':')} was the right one")
            break 
```
