The key is that the "see" value will always be in front of the value we are working with.  

[The Look & Say sequence is so cool](https://www.youtube.com/watch?v=koHqD2s0SdM) by [Dr. Trefor Bazett](https://www.youtube.com/@DrTrefor) on youtube was used to get my head around it.

I made the input into a string so i can iterate through it and check what i need

```python
## Created a function i could call severeal times in a for loop as i need ot run it 40 times for the sequence
def create_look_and_say_sequence(sequence_input : str) -> str:
    
    ## Using a "dirty for loop" agian as i cant change iterator in a foor loop like i can in c++ during runtime.
    i = 0
    the_next_sequence = ""  ## We are tryign to make a new sequence also in a string
    while i < len(sequence_input):
        current_number = sequence_input[i]
        how_many = 0  ## This is how many times a char has showed up
        #print(f"going to index {i}")
        for j in range(i, len(sequence_input)):
            if current_number == sequence_input[j]:
                how_many += 1
            else:
                break ## We break out early if the type of number changes
        the_next_sequence += str(how_many)+current_number 
        i = i+how_many ## I thought this was clever using the number to move the iterator forward.
    return the_next_sequence
```

```python
## used this as my test to see if it fit with what Dr. Trefor Bazett showed in his video 
input_test = "3"
print(input_test)
for i in range(7):
    input_test = create_look_and_say_sequence(input_test)
    print(input_test)

print(f"13211321322113 is the final sequence from the video")
```
