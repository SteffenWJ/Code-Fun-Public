```python
## Tiny little helper function, onyl used in Part 2
def is_positon_in_dict(pos_dict,x_pos,y_pos):
    test_position = (x_pos,y_pos)
    if test_position in pos_dict:
        pos_dict[test_position] =+1
    else:
        pos_dict[test_position] = 1

    return pos_dict
```

```python
def track_the_houses(text_line):
    ## I will use a dict to check if a cordinate set exists, if not it will be added with a value of 1.
    ## If the cordiantes exists, 1 is added to the value. This is more so i can follow it what house gets what and suspect it might be in part 2
    ## The list of the path is also added, but that is more for me as i wante to make a map of the path as it could be funny.
    
    x_pos = 0    # East and West
    y_pos = 0    # North and South

     #Start at 0.0 and deliver a present
    pos_list = [[0,0]]
    house_dict = {(0,0) : 1}


    for i in range(len(text_line)):
        text_line[i]
        ## Updates the postion
        if text_line[i] == '^': ## North
            y_pos +=1
        elif text_line[i] == 'v': ## South
            y_pos -= 1
        elif text_line[i] == '>': ## East
            x_pos += 1
        elif text_line[i] == '<': ## West
            x_pos -= 1
        
        pos_list.append([x_pos,y_pos])
        test_position = (x_pos,y_pos)
        if test_position in house_dict:
            house_dict[test_position] =+1
        else:
            house_dict[test_position] = 1

    return house_dict, pos_list ## Only need the length of the house_dict
```

```python
with open(input_path,'r') as text:
    line = text.readline()
    dict_houses, positions_list =  track_the_houses(line)
    

print(f"The number of houses to get atleast one present is {len(dict_houses)}")
```

```python
## Day 3 Part 2
def track_the_houses_ROBO(text_line):
    ## Is a bit of the same as before.
    ## Can use the same methode, but isntead of only trackign Santa we track both.
    ## Modfying the orginal function

    x_pos_santa = 0
    y_pos_santa = 0

    x_pos_robot = 0
    y_pos_robot = 0

     #Start at 0.0 and they both deliver a present
    house_dict = {(0,0) : 2} #HEHE


    for i in range(len(text_line)):
        text_line[i]
        ## Updates the postion
        y_pos = 0
        x_pos = 0
        if text_line[i] == '^': ## North
            y_pos = 1
        elif text_line[i] == 'v': ## South
            y_pos = -1
        elif text_line[i] == '>': ## East
            x_pos  = 1
        elif text_line[i] == '<': ## West
            x_pos = -1

        if i % 2 == 0:
            ## Santa move
            x_pos_santa += x_pos
            y_pos_santa += y_pos
            house_dict = is_positon_in_dict(house_dict,x_pos_santa,y_pos_santa)
        else:
            ## Robot move
            x_pos_robot += x_pos
            y_pos_robot += y_pos
            house_dict = is_positon_in_dict(house_dict,x_pos_robot,y_pos_robot)

    return house_dict ## Only need the length of the house_dict
```

```python
with open(input_path,'r') as text:
    line = text.readline()
    dict_houses =  track_the_houses_ROBO(line)
    

print(f"The number of houses that Santa and ROBO-Santa gave atleast one present was: {len(dict_houses)}")
```
