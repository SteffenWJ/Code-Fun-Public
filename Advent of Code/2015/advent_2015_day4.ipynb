```python
import hashlib
```

```python
## This is the only input for now
input_text = ""
```

```python
## Used this genereate the md5 hash
def generate_md5_hash(test_input):
    md5 = hashlib.md5()
    md5.update(test_input.encode('utf-8'))
    return md5.hexdigest()
```

```python
## This is not the optimal way of doign it as it is brute force method, but works.
## A smarter way would be to check more spcific values as there is only so many possible combinations that could potentioaly
## Have 0 in front of the rest of the value.
def find_the_value(input_string : str, goal : str = "00000"):
    start_text = input_string
    int_value = 0

    current_test = None
    current_output = None
    while(True):
        current_test = start_text+str(int_value)
        current_output = generate_md5_hash(current_test)

        if current_output[0:len(goal)] == goal:
            return current_test, current_output
        else:
            int_value +=1
    
```

```python
## Day 5 part 1
hash_output, hex_output = find_the_value(input_text, "00000")
print(f"The Has output of {hash_output}: gives us the hex value of {hex_output[0:10]}... : with five 0")
## Day 5 part 2 ## No real change i could use the same code, but with the different wanted 0
hash_output, hex_output = find_the_value(input_text, "000000")
print(f"The Has output of {hash_output}: gives us the hex value of {hex_output[0:10]}... : with six 0")
```
