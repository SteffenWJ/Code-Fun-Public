```python
## Day 8 part 1

def get_hex_value(hexx_chars : str):
    ascii_output = ""  ## I want the ascii string in case is needed for part 2
    ##print(f"Getting HEX: {hexx_chars}")
    for i in range(0, len(hexx_chars), 2): ## It needs to be this way.
        part = hexx_chars[i : i + 2]
        ch = chr(int(part, 16))
        ascii_output += ch

    return ascii_output

## I could skip the gathering of the charcters and only look at the differences as that would still be the correct answer.
def get_charter_string(text_line: str):
    value_string = [] ## I am gathering the chars in the string
    full_string = [] ## The reason we also get the full string is becaus of the way some chars work, like the hex decimal will be 2 chars, but will only show as 1 in the length 

    ##for i in range (len(text_line)): ## Orginaly used a for loop, but found out you can change the iterator in python like you can in C++
    i = 0
    while i < len(text_line):
        current_charcter = text_line[i] 
        if current_charcter == '"':
            ## keep quotes in full_string, but not in value_string
            full_string.append(current_charcter)
            i += 1

        elif current_charcter == '\\':
            if (i + 1 < len(text_line) and text_line[i+1] == 'x' and i + 3 < len(text_line)  # we need 2 chars after x
                and all(c in string.hexdigits for c in text_line[i+2:i+4])): ## This was need as i had to check if the chars where hex digit.
                # Converting to get the Hex Value
                value_string.append(get_hex_value(text_line[i+2:i+4]))
                # Add the String to the full onbe
                full_string.extend(text_line[i:i+4])
                i += 4 

            elif i + 1 < len(text_line) and text_line[i+1] == '"':
                value_string.append('"')
                full_string.append('\\')
                full_string.append('"')
                i += 2

            elif i + 1 < len(text_line) and text_line[i+1] == '\\':
                value_string.append('\\')
                full_string.append('\\')
                full_string.append('\\')
                i += 2

            else:
                value_string.append('\\')
                full_string.append('\\')
                i += 1

        elif current_charcter == '\n':
            ## I am skiping the new line, as it will be in the string, but not part of the difference calcualtions
            i += 1

        else:
            value_string.append(current_charcter)
            full_string.append(current_charcter)
            i += 1

    return value_string, full_string
```

```python
## Day 8 Part 1.
char_strings = []
full_strings = []

sum_chars = 0
sum_strings = 0
with open(input_path,'r') as text:
    for line in text:
        char_strings, full_strings = get_charter_string(line)
        sum_strings += len(full_strings)
        sum_chars += len(char_strings)

print(f"Full string - Char only string : {sum_strings} - {sum_chars} = {sum_strings-sum_chars}")
```

```python
def get_bigger_string(text_line: str):

    text_line = text_line.rstrip('\n')

    full_string = []      # original code chars
    bigger_string = []    # encoded chars

    for ch in text_line:
        full_string.append(ch)

    bigger_string.append('"') 

    for ch in text_line:
        if ch == '\\':
            # escape backslash
            bigger_string.append('\\')
            bigger_string.append('\\')
        elif ch == '"':
            # escape quote
            bigger_string.append('\\')
            bigger_string.append('"')
        else:
            bigger_string.append(ch)

    bigger_string.append('"')

    return bigger_string, full_string
```

```python
sum_full = 0      # sum of encoded lengths
sum_strings = 0   # sum of original code lengths

with open(input_path, 'r') as text:
    for line in text:
        bigger_string, full_strings = get_bigger_string(line)

        # lengths per line
        len_bigger = len(bigger_string)
        len_original = len(full_strings)

        ## print(f"Bigger (bigger): {len_bigger}  Original: {len_original}")
        ## print(bigger_string) 

        sum_full += len_bigger
        sum_strings += len_original

print(f"Total Encoded - Total Original : {sum_full} - {sum_strings} = {sum_full - sum_strings}")
```
