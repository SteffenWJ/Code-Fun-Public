```python
input_path = ""
```

```python
## Creating a simple dict to contain the info
def create_raindeer_dict(raindeer_name: str, fly_speed :int , distance_time : int, rest_time : int,  the_dict: dict):
    the_dict[raindeer_name] = {"speed": fly_speed, "flying" : distance_time, "rest" : rest_time}
```

```python
## Added a point system to the dict, it is the same as before, but makes it simpler to keep using the same dict
def create_raindeer_dict_extended(raindeer_name: str, fly_speed :int , distance_time : int, rest_time : int,  the_dict: dict):
    the_dict[raindeer_name] = {"speed": fly_speed, "flying" : distance_time, "rest" : rest_time, "distance" : 0, "points": 0}
```

```python
raindeer_dict = {}
raindeer_set = set() ## I like sets....

with open(input_path, 'r') as text:
    for line in text:
        line_split = line.split()
        name_raindeer = line_split[0]
        fly_speed = int(line_split[3])
        flying_time = int(line_split[6])
        rest_time = int(line_split[13])
        #create_raindeer_dict(name_raindeer,fly_speed,flying_time,rest_time,raindeer_dict)
        create_raindeer_dict_extended(name_raindeer,fly_speed,flying_time,rest_time,raindeer_dict)
        print(f"Name: {name_raindeer}, speed: {fly_speed}, flying {flying_time}, rest {rest_time}") ## Remove if the input is to big, but in this case it was not.
        raindeer_set.add(name_raindeer)
```

```python
def longest_distance_travled(raindeer_dict : dict, raindeer_set: set , race_time : int):
    best_raindeer = ""
    best_distance = 0
    for raindeer in raindeer_set:
        current_distance = 0

        for t in range(1, race_time+1): ## Was off by one
            flying_cycle = raindeer_dict[raindeer]["flying"] + raindeer_dict[raindeer]["rest"]
            t_in_cycle = (t - 1) % flying_cycle
            
            if t_in_cycle < raindeer_dict[raindeer]["flying"]:  # in the flying part of the cycle
                current_distance += raindeer_dict[raindeer]["speed"]
        
        if current_distance > best_distance:
            best_distance = current_distance
            best_raindeer = raindeer
        
    return best_raindeer, best_distance
```

```python
## Day 14 part 1
raindeer_name,the_distance = longest_distance_travled(raindeer_dict,raindeer_set,2503)
print(f"Best raindeer was {raindeer_name} with a distance off {the_distance}")
```

```python
## Simple helper to make tihngs easier,
def calculate_distance(raindeer_dict : dict, raindeer_name : str, time_step : int):
    flying_cycle = raindeer_dict[raindeer_name]["flying"] + raindeer_dict[raindeer_name]["rest"]
    t_in_cycle = (time_step - 1) % flying_cycle
    if t_in_cycle < raindeer_dict[raindeer_name]["flying"]:
        raindeer_dict[raindeer_name]["distance"] += raindeer_dict[raindeer_name]["speed"]
```

```python
## It is moreo r less the same task switched around so instead of simulating on raindeer at a time isntead we simualte the race and every secound we update the time.
def parralel_race(raindeer_dict : dict, raindeer_set: set , race_time : int):
    for t in range(1, race_time+1):
        for name in raindeer_set:
            calculate_distance(raindeer_dict,name,t)
        
        max_dist = max(raindeer_dict[name]["distance"] for name in raindeer_set)
        for name in raindeer_set:
            if raindeer_dict[name]["distance"] == max_dist: ## So a issue i had here is only one of the raindeer would get a point so i added that every raindeer that hits the max distance gets a poitn for the time tic
                raindeer_dict[name]["points"] += 1

    most_points_name = max(raindeer_dict, key=lambda r: raindeer_dict[r]["points"])
    most_points_value = raindeer_dict[most_points_name]["points"]
    return most_points_name, most_points_value
```

```python
## Day 14 part 2
raindeer_name, the_points = parralel_race(raindeer_dict, raindeer_set, 2503)
print(f"Best raindeer was {raindeer_name} with {the_points} points")
```
