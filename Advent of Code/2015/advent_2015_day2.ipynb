The input will be
length x width x heigh

Calculations of needed paper: 2\*l\*w + 2\*w\*h + 2\*h\*l = Value
And the "Slack" is the AREA of the smallets side. 

```python
## Day 2 part 2
## Simple enough. Iterate each line, split by 'x' and then calcualte it and look for the smalest value and calcualte the area
def calculate_wraping_paper(text_line : str, verbose : bool = False) -> int:
    length_val, width_val, height_val = list(map(int,text_line.split('x'))) #I know the list(map(type,list_object)) looks wierd.
    ## I am going to make the calculations slit up
    length_area = 2*length_val*width_val ## 2*l*w
    width_area = 2*width_val*height_val  ## 2*w*h
    height_area = 2*height_val*length_val ## 2*h*l

    slack_list = [length_val*width_val,length_val*height_val,width_val*height_val]   #Area of a Square Side * Side
    slack_area = min(slack_list)

    #One could use Numpy or ohter things to make it a bit fanciser looking
    final_area = length_area+width_area+height_area+slack_area

    if verbose:
        print(f"Orginal input: {text_line.strip()}")
        print(f"Length: {length_val}\nWidth: {width_val}\nHeight: {height_val}")
        print(f"Area was: {final_area} = {length_area}+{width_area}+{height_area}+{slack_area}")

    return final_area


```

```python
with open(input_path,'r') as text:
    #count = 0
    sum_wrapping_paper = 0
    for line in text:
        #print(line)
        sum_wrapping_paper += calculate_wraping_paper(line,False)
    #    count += 1
    #    if count == 5:
    #        break

print(f"The sum of wrapping paper is: {sum_wrapping_paper}")
```

```python
def calculate_ribbon(text_line : str, verbose : bool = False) -> int:
    value_lust = list(map(int,text_line.split('x')))
    length_val, width_val, height_val = value_lust 

    low, mid, _ = sorted(value_lust)

    ## Ribbon We need the smallest distance. So width is locked so the smallest of length of height is needed
    ## So above there is a error i am leaving it to remind my self. WIDTH of the Ribbon is set, but it is still need all three sides including WIDTH of the box, clever Puzzle (One day i will earn to read)
    ## So find which TWO sides are the smallest like before.

    ## I will sort the values and then take the two smallest ones....
    ribbon_dim = low+low+mid+mid
    bow_dim = length_val*width_val*height_val
    final_ribbon = ribbon_dim+bow_dim

    if verbose:
        print(f"Orginal input: {text_line.strip()}")
        print(f"Length: {length_val}\nWidth: {width_val}\nHeight: {height_val}")
        print(f"Smalest of Length: {length_val} vs Height: {height_val}")
        #print(f"Ribbon Formula was| Ribbon Dimmensions w+w+{which_side}+{which_side} and bow: l*w*h ")
        print(f"Ribbon should be : {final_ribbon} = {ribbon_dim}+{bow_dim}")

    return final_ribbon
```

```python
with open(input_path,'r') as text:
    #count = 0
    sum_ribbon = 0
    for line in text:
        #print(line)
        sum_ribbon += calculate_ribbon(line,False)
    #    count += 1
    #    if count == 5:
    #        break

print(f"needed Ribon was was: {sum_ribbon}")
## To large 3788504
```
