```python
from itertools import permutations
import math
```

```python
## helper functions that are not longer used. Made for PArt 1
## This is so i can create the needed key from different parts.
## As i look at the input, this is not needed, but i thought it might be some dual entries.
def create_the_key(place_A  : str, place_B : str):
    return tuple(sorted([place_A,place_B]))

## These two are mostyl to handle duplicates in the input
def create_the_distance_dict_old(place_A : str, place_B : str, distance : str, the_dict: int):
    the_key = create_the_key(place_A,place_B)
    if the_key not in the_dict:
        the_dict[the_key] = distance

def create_places_list(place_A: str, places_list: list):
    for name, _ in places_list:
        if name == place_A:
            return  ## it is alread
    places_list.append(place_A)
```

```python
## These helper functions are used in Part 1
def create_distance_dict(place_A : str, place_B : str, distance : int, the_dict: dict):
    ## Create the two keys for each entry.
    ## I did this becaus i want both dicts to have the values in them. 
    if place_A not in the_dict:
        the_dict[place_A] = []
    if place_B not in the_dict:
        the_dict[place_B] = []
    
    ## Checking if the values are already added to the dict, it is not needed for this input, i overthought it.
    if not any(n == place_B for n, d in the_dict[place_A]):
        the_dict[place_A].append((place_B, distance))

    if not any(n == place_A for n, d in the_dict[place_B]):
        the_dict[place_B].append((place_A, distance))

## This only gets the distance between two places from a dict. 
def get_the_distance(place_A : str, place_B : str, the_dict : dict) -> int:
    for neighbor, dist in the_dict[place_A]:
        if neighbor == place_B:
            return int(dist) ## Rember to convert the type to int as for the dict it will be a string.
    return None
```

```python
input_path = ""
```

Now we have the dict and List, time to get to the traversal

```python
travel_dict = {}
set_of_places = set()
count = 0

## First lets get the different parts we need
with open(input_path, 'r') as text:
    for line in text:
        #print(count)
        place_A,_,place_B,_,distance = line.split()
        create_distance_dict(place_A,place_B,distance,travel_dict)
        set_of_places.add(place_A)
        set_of_places.add(place_B)
```

```python
def find_the_best_distance(the_places_dict : dict ,the_set_of_places : set, what_distance : str = "Shortest"):
    ## A function created after i read part 2 as only a few things had to be changed to i made it into a function
    
    ## Sets the targets to the infinitley high or infenetely low
    if what_distance.upper() == "SHORTEST":
        target_distance = math.inf
    elif what_distance.upper() == "LONGEST":
        target_distance = -math.inf
    else:
        print(f"The distance type of {what_distance} could not be found, only use Longest or Shortest")
        return None, None


    for place in permutations(set_of_places):  ## Permutations is really neet as it creates EVERY possible version of the places in the List.
        sum_distance = 0

        for i in range(0, len(place)-1):
            place_A = place[i]
            place_B = place[i+1]
            distance = get_the_distance(place_A, place_B, travel_dict)
            sum_distance += distance

        if what_distance.upper() == "SHORTEST":
            if sum_distance < target_distance:
                target_distance = sum_distance
                best_route = place
        elif what_distance.upper() == "LONGEST":
            if sum_distance > target_distance:
                target_distance = sum_distance
                best_route = place

    return best_route, target_distance
```

```python
## Day 9 part 1 - shortest
best_route, best_distance = find_the_best_distance(travel_dict,set_of_places)
print(f"The best shortest route with a distance of {best_distance} was: {best_route}")
```

```python
## Day 9 part 2 - longest
best_route, best_distance = find_the_best_distance(travel_dict,set_of_places,"longest")
print(f"The best longest route with a distance of {best_distance} was: {best_route}")
```
