```python
input_path = ""
```

Rules:

Atleast 3 Vowels of the following [a,e,i,o,u]
Contains atleast 1 letter that appears twice in a row
dose not contain strings contaning [ab,cd,pq,xy]

```python
## A helper function to check if it is a naughty
def checking_if_naughty(test_input):
    naughty_list = ["ab","cd","pq","xy"]
    for naughty in naughty_list:
        if test_input == naughty:
            return True
        else:
            return False
```

```python
## Going to make a cascading failure thing.
def is_the_string_nice(text_input):
    number_of_vowels = 0
    double_letters = False

    #naughty_list = ["ab","cd","pq","xy"]
    #nice_list = ['a','e','i','o','u']

    for i in range(len(text_input)-1):
        ## Check if naughty
        if text_input[i]+text_input[i+1] in ("ab","cd","pq","xy"):
            return False
        elif text_input[i] == text_input[i+1]:
            double_letters = True

        if text_input[i] in ('a','e','i','o','u'):
            number_of_vowels += 1


    if double_letters == True and number_of_vowels >= 3: ## Rember to point the operator in the right direction....
        return True    
    return False
```

```python
with open(input_path,'r') as text:
    number_of_nice = 0
    for line in text:
        if is_the_string_nice(line):
            number_of_nice += 1

print(f"The Number of Nice entries {number_of_nice}")
```

```python
## Day 5 part 2
## Updating the Cascading Failure
def is_the_string_nice_new(text_input):
    pair_of_two_letters = False
    double_letters = False

    ## it only has nice triggers
    for i in range(len(text_input)-1):
        ## Check if naughty
        if text_input[i]+text_input[i+1] in text_input[i+2:] and not pair_of_two_letters:
            pair_of_two_letters = True

        if i+2 < len(text_input):
            if text_input[i] in text_input[i+2]:
                double_letters = True

        ## Early break out if both are true
        if pair_of_two_letters and double_letters:
            return True
    return False
```

```python
with open(input_path,'r') as text:
    number_of_nice = 0
    for line in text:
        if is_the_string_nice_new(line):
            number_of_nice += 1

print(f"The Number of Nice entries {number_of_nice}")
```
