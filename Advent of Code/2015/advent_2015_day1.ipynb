```python
## Input your own path to your input here
input_path = "path_to_input.txt"
```

```python
## Counts each, could do this with unquie and count.
def get_paranthesis_split_count(text_line):
    start_paranthese = 0
    end_paranthese = 0
    for i in range(len(text_line)):
        if text_line[i] == '(':
            start_paranthese += 1
        elif text_line[i] == ')':
            end_paranthese += 1
    return start_paranthese, end_paranthese
```

```python
## Day 1
## It is a simple count every thing with '(' minus ')'
with open(input_path,'r') as text:
    line = text.readline()
    start_paranthese, end_paranthese = get_paranthesis_split_count(line)

dif = start_paranthese-end_paranthese
print(f"Start: {start_paranthese} : End {end_paranthese} = {dif}")
```
