```python
from itertools import permutations
```

```python
input_path = ""
```

```python
## Used same approch from Day 9
def create_seating_dict(person_A : str, person_B : str, gain_lose : str, happiness : int, the_dict: dict):
    #print(f"{person_A} next to {person_B} {gain_lose} {happiness}")
    signed_happiness = happiness if gain_lose == "gain" else -happiness
    if person_A not in the_dict:
        the_dict[person_A] = {}
    the_dict[person_A][person_B] = signed_happiness
```

```python
people_dict = {}
set_of_people = set() ## The set is used so i get unique people

with open(input_path, 'r') as text:
    for line in text:
        line_split = line.split()
        create_seating_dict(line_split[0], line_split[10].strip('.'), line_split[2], int(line_split[3]), people_dict)
        set_of_people.add(line_split[0])
        set_of_people.add(line_split[10].strip('.'))
```

```python
def total_happiness(seating, people_dict):
    sum_happiness = 0

    for i in range(len(seating)):
        person_A = seating[i]
        person_B = seating[(i + 1) % len(seating)]  ## We wrap around with this
        sum_happiness += people_dict[person_A][person_B]
        sum_happiness += people_dict[person_B][person_A]

    return sum_happiness

def calcualte_best_seating(set_of_people : set, people_dict : dict):
    people_in_set = list(set_of_people)
    first_seat = people_in_set[0]  ## By picking a single person first then it is easier to set the rest.
    ohter_seats = people_in_set[1:] ## Slice SLICE

    best_score = None
    best_seating = None

    for perm in permutations(ohter_seats): ## Uses thies to get ever possible combination.
        seating = (first_seat,) + perm 
        score = total_happiness(seating, people_dict)

        if best_score is None or score > best_score:
            best_score = score
            best_seating = seating

    return best_score, best_seating
```

```python
## Day 13 part 1
happiness, best_seatings = calcualte_best_seating(set_of_people,people_dict)

print(f"Highest happiness was: {happiness}")
print(f"Best Seating would then be: {best_seatings}") 
```

Day 13 part 2 : simplest and easiest way would be to add my self for each person in the existing dict

```python
## To add myself to the data i will use what i have and the set of people.
for person in set_of_people:
    create_seating_dict(person,"myself","lose", 0 ,people_dict) ## Lose and gain dose not mater in the case of myself
    create_seating_dict("myself",person,"lose", 0 ,people_dict)
    #print(person)
set_of_people.add("myself")
```

```python
## Day 13 part 2
happiness, best_seatings = calcualte_best_seating(set_of_people,people_dict)

print(f"Highest happiness was: {happiness}")
print(f"Best Seating would then be: {best_seatings}")
```
