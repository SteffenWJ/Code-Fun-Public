```python
## Function made for day 2 part 1:

## The idea is that any row in from the input is processed through a check list, and if there is any triggers it will return False, being a unsafe route.
def is_it_safe(value_array):
    ## Change is used to monitor if the row is incrementing or decreasing.
    change = None
    for i in range(len(value_array)-1): # The for loop is run for one less then the length of the row given.
        dif = value_array[i]-value_array[i+1]
        #print(dif)
        if dif == 0 or abs(dif) > 3 :  ## First check is the difference between neighbouring numbers and if they are 0 or the difference betwen them is to large.
            return False
        ## next the changes is monitored. First time it triggers it sets the type, and if triggered agian it checks if it has changed its projectory.
        elif dif > 0 :
            if change == None:
                change = '-'
            elif change == '+':
                return False
        elif dif < 0 : 
            if change == None:
                change = '+'
            elif change == '-':
                return False
    return True 
```

```python
## Function for part 2 where a single problem in in a row that would set it as unsafe can be "REMOVED", the key here is removed. 
def is_it_safe_dampener(value_array):
    # If it is safe in the first pass it is let through. Else the row is processed.
    safe = is_it_safe(value_array)
    if safe:
        return True
    
    # This is a ugly brute force way, but it works.
    for i in range(len(value_array)):
        # Each value is removed one by one to see if there is a safe route for the ROW.
        new_array = np.delete(value_array,i)
        safe = is_it_safe(new_array) 
        if safe: ## If a single safe route is found, we exit the loop and return True, as we are not intrested in the "BEST" or nay thing, onyl that there is a safe route.
            return True
    return False
```

```python
## Day 2 - Part 1
count = 0
with open(input_path, 'r') as text:
#with open("test_day_2.txt", 'r') as text: ## Test day 2 is a input created by me from the example from AOC 2024 for sanity.
    for line in text:
        test = line.split(" ")
        B = np.array(test,dtype=int)
        #print(B)
        #print(f"It was safe: {is_it_safe(B)}")
        safe= is_it_safe(B)
        ## To find out how many safe rows there is, a simple count was added.
        if safe:
            count += 1
print(f"Number of safe: {count}") 
```

```python
## Day 2 - Part 2
## Same as part one, everything new is in the is_it_safe_dampener function
count = 0
with open(input_path, 'r') as text:
#with open("test_day_2.txt", 'r') as text:
    for line in text:
        test = line.split(" ")
        B = np.array(test,dtype=int)
        #print(B)
        if is_it_safe_dampener(B):
            count += 1

print(f"Number of safe with Dampener: {count}")
```
