```python
## Day 1 - Part 1

## First i loaded in the data, used numpy as i know it as a lot of strong math uses.
all_input = np.loadtxt(input_path,dtype=int)

#Small Sanity check
sanity_all = np.array([[18944,47230],[94847,63037],[93893,35622],[37174,43365],[77982,51397]])
print(sanity_all)
print("should be Equal:")
print(all_input[:5])
```

```python
## More Sanity checks to be certain the library performes as i expected
left_san,right_san = np.hsplit(sanity_all,2)
left_col,right_col = np.hsplit(all_input,2)

print(left_san)
print(left_col[:5])

print("")

print(right_san)
print(right_col[:5])
```

```python
## As the task was to get the difference between the lowers number in one colum agianst the ohter, the easiet way to do this is to sort each.
left_sort = np.sort(left_col,axis=None)
right_sort = np.sort(right_col,axis=None)
## Absulute value was taken as we are looking for the difference between the two value colum.
difference = abs(left_sort-right_sort)

# This was a quick test to be certain i had not missed any thing
for i in range(5):
    print(f"Difference: {left_sort[i]} - {right_sort[i]} = {difference[i]}")

## The difference is summed.
dif_sum = difference.sum()
print(dif_sum)
print(f"Difference was {dif_sum}") # Any correct answer is more so if i play around i dont have to recheck Advent of Code for what my answer was.
```

```python
## Now putting it to work.

## A good thing with unique from numpy was that it could return each unique value from a list and how many times this value was in the list.
left_unique,left_count = np.unique(left_sort,return_counts=True)
right_unique,right_count = np.unique(right_sort,return_counts=True)

## This information is then made into a dict for easy acces.
left_dict = dict(zip(left_unique,left_count))
right_dict = dict(zip(right_unique,right_count))
```

```python
## used a list for my difference instead of numpy array as i like it better for appending data in a for loop
dif = []
for i in range(len(left_sort)): #len(left_sort)
    # Get left number
    # Check if it exists in right
    # IF yes * number in right
    # IF no * 0
    # sum result.

    ## Used get as as it allowed to have a defult value of a key in the dict did not exist and set it to zero, so made it easy to multiply as needed.
    multi = right_dict.get(left_sort[i],0)
    dif.append(left_sort[i]*multi)

## Finaly sum every the difference.
print(f"Answer to Day 1 part 2 should be: {sum(dif)}")
```
