```python
def generate_hashtag(s):
    if len(s) == 0:
        return False
    text_chunks = "#"
    for text_part in s.split(" "):
        text_chunks += text_part.capitalize()

    if len(text_chunks) > 140:
        return False
    
    return text_chunks
```

an equals method, to check that two vectors that have the same components are equal  

Note: the test cases will utilize the user-provided equals method.

```python
from math import sqrt 

class Vector:
    def __init__(self, the_list):
        self.list = the_list

    def __is_equal_length(self, ohter_vector):
        ## Checks if the lists are equal in length
        if len(self.list) != len(ohter_vector.list):
            raise ValueError("Lists are different sizes")
        
    def __is_Vector_type(self, ohter_vector):
        if not isinstance(ohter_vector, Vector):
            raise TypeError("Expects a Vector object")
    
    def add(self, ohter_vector):
        self.__is_Vector_type(ohter_vector)
        self.__is_equal_length(ohter_vector)

        return Vector([self.list[i] + ohter_vector.list[i] for i in range(len(self.list))])
    
    def subtract(self, ohter_vector):
        self.__is_Vector_type(ohter_vector)
        self.__is_equal_length(ohter_vector)

        return Vector([self.list[i] - ohter_vector.list[i] for i in range(len(self.list))])
    
    def dot(self, ohter_vector):
        self.__is_Vector_type(ohter_vector)
        self.__is_equal_length(ohter_vector)

        return sum([self.list[i] * ohter_vector.list[i] for i in range(len(self.list))])
    
    def norm(self):
        return sqrt(sum(x * x for x in self.list))
    
    def equals(self, ohter_vector):
        return isinstance(ohter_vector, Vector) and self.list == ohter_vector.list
    
    def __str__(self):
        return f"({','.join(str(x) for x in self.list)})"
    

a = Vector([1, 2, 3])  
b = Vector([3, 4, 5])  
c = Vector([5, 6, 7, 8])  

test = a.add(b)      # should return a new Vector([4, 6, 8])  
print(test.list)
test = a.subtract(b) # should return a new Vector([-2, -2, -2])
print(test.list)
test =  a.dot(b)      # should return 1*3 + 2*4 + 3*5 = 26
print(test)  
test = a.norm()      # should return sqrt(1^2 + 2^2 + 3^2) = sqrt(14)
print(test)  
#a.add(c)      # raises an exception  
```

Define a function that takes in two non-negative integers a and b and returns the last decimal digit of a^b.  Note that a and b.

For example, the last decimal digit of 9^7 is 9 since 9^7 = 4782969.

pow(7,4,10) creates this
7^4 = 2401  
2401 % 10 = 1  
Built in fuction in python

```python
def last_digit(n1, n2):
    return pow(n1,n2, 10)

print(last_digit(9,7))
print(last_digit(7,4))
```

Write a function that when given a URL as a string, parses out just the domain name and returns it as a string. For example:

* url = "http://github.com/carbonfive/raygun" -> domain name = "github"
* url = "http://www.zombie-bites.com"         -> domain name = "zombie-bites"
* url = "https://www.cnet.com"                -> domain name = cnet"

```python
## This is dumb, but it works. A better way would be to split the string and be a bit smart
def domain_name(url):
    start_point = 0
    if "http://www." in url:
        start_point = 11
    elif "https://www." in url:
        start_point = 12
    elif "https://" in url:
        start_point = 8
    elif "http://" in url:
        start_point = 7
    elif "www." in url:
        start_point = 4
    index = start_point+1
    while(index < len(url)):
        if url[index] == "/":
            break
        elif url[index] == ".":
            break
        index += 1
    ##print(url[start_point:index])
    return url[start_point:index]
```

First non-repating charcter  


Write a function that takes a string input, and returns the first character that is not repeated anywhere in the string.

For example, if given the input "stress", the function should return 't', since the letter t only occurs once in the string, and occurs first in the string.

As an added challenge, upper- and lowercase characters are considered the same character, but the function should return the correct case for the initial character. For example, the input "sTreSS" should return "T".

If a string contains only repeating characters, return an empty string ("");

Note: despite its name in some languages, your function should handle any Unicode codepoint:

```python
def first_non_repeating_letter(s):
    length_of_string = len(s)
    pointer_index = 0
    candidate = ""
    while pointer_index < length_of_string:
        candidate = s[pointer_index]

        if s.upper().count(candidate.upper()) == 1:
            break

        pointer_index += 1
    else:
        return ""

    return candidate

test_string = "lalalarattasaaaattttaaaa"
print(first_non_repeating_letter(test_string))
```
