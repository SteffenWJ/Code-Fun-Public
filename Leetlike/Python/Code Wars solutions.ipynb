**Multiples of 3 or 5**

If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.  
Finish the solution so that it returns the sum of all the multiples of 3 or 5 below the number passed in.  
Additionally, if the number is negative, return 0.    
Note: If a number is a multiple of both 3 and 5, only count it once.  

Courtesy of projecteuler.net ([Problem 1](https://projecteuler.net/problem=1))

```python
def solution_multiples_of_3_5(number):
    list_numbers = []
    for i in range(number):
        if i % 3 == 0 and i != 1:
            list_numbers.append(i)
        elif i % 5 == 0  and i != 1:
            list_numbers.append(i)
    
    return sum(list_numbers)

print(solution_multiples_of_3_5(100))
```

Bit Counting

Write a function that takes an integer as input, and returns the number of bits that are equal to one in the binary representation of that number. You can guarantee that input is non-negative.

Example: The binary representation of 1234 is 10011010010, so the function should return 5 in this case

```python
def count_bits(n):
    ## First lets get it to bit
    bit_string = format(n,"016b") ## Set it to a 16 bit string
    count = 0
    for bit in bit_string:
        if bit == '1':
            count += 1
    
    return count

print(count_bits(1234))
```

Your job is to create a calculator which evaluates expressions in Reverse Polish notation.  

For example expression 5 1 2 + 4 * + 3 - (which is equivalent to 5 + ((1 + 2) * 4) - 3 in normal notation) should evaluate to 14.  
For your convenience, the input is formatted such that a space is provided between every token.  
Empty expression should evaluate to 0.  
Valid operations are +, -, *, /.

You may assume that there won't be exceptional situations (like stack underflow or division by zero).

[Wikipage About Reverse Polish Notation](https://en.wikipedia.org/wiki/Reverse_Polish_notation)  

The Explanation is not great, but every time we hit a perator we use the LAST two numbers added. That is why 5 1 2 + becomes 5 + (1 + 2) as the (1+2) is tha last. That is also why we then have 5+(3*4)

```python
test = "5 1 2 + 4 * + 3 -"
def calc(expr):
    if len(expr) == 0:
        ## A qucik check if the string is empty
        return 0.0
    
    stack = []

    value_first = 0
    value_last  = 0
    for char_in_string in expr.split():                 ## The char needs to be split up into its different individual parts from the string.
        if char_in_string not in '+-*/':               ## If it is not one of the operators we add the number to the stack
            stack.append(float(char_in_string))
        else:
            value_last  = stack.pop()                       ## We go backwards becaus of how the operators are added to the stack if we have 5 1 2 and we pop then we get 2 not 
            value_first = stack.pop()
            if char_in_string == '+':
                stack.append(value_first + value_last)
            elif char_in_string == '-':
                stack.append(value_first - value_last)
            elif char_in_string == '*':
                stack.append(value_first * value_last)
            elif char_in_string == '/':
                stack.append(value_first / value_last)
    return stack[0]
print(calc(test), "Should be 14")
```

Input

Output

In this kata you will create a function that takes a list of non-negative integers and strings and returns a new list with the strings filtered out.

filter_list([1,2,'a','b']) == [1,2]  
filter_list([1,'a','b',0,15]) == [1,0,15]  
filter_list([1,2,'aasf','1','123',123]) == [1,2,123]  

```python
def filter_list_easier_read(l):
    ## Easier read, but slower, but functionaly the same
    new_list = []

    for char in l:
        if not isinstance(char,str):
            new_list.append(char)
    return new_list

def filter_list(l): ## More correct Python and faster
    return [x for x in l if not isinstance(x,str)]


test_list_A = ([1,2,'a','b'])
test_list_B = ([1,'a','b',0,15])
test_list_C = ([1,2,'aasf','1','123',123])

print(f"Test A : {filter_list_easier_read(test_list_A)} : should be [1, 2]")
print(f"Test B : {filter_list_easier_read(test_list_B)} : should be [1, 0, 15]")
print(f"Test C : {filter_list_easier_read(test_list_C)} : should be [1, 2, 123]")

print(f"Test A : {filter_list(test_list_A)} : should be [1, 2]")
print(f"Test B : {filter_list(test_list_B)} : should be [1, 0, 15]")
print(f"Test C : {filter_list(test_list_C)} : should be [1, 2, 123]")
```

Descending Order

Your task is to make a function that can take any non-negative integer as an argument and return it with its digits in descending order. Essentially, rearrange the digits to create the highest possible number.

Input: 42145 Output: 54421  
Input: 145263 Output: 654321  
Input: 123456789 Output: 987654321  

```python
def descending_order(num):
    temp_list = sorted([int(x) for x in str(num)],reverse=True)
    reverse_number = int("".join(str(j) for j in temp_list))
    return reverse_number

test_A = 42145
test_B = 145263
test_C = 123456789

print(f"Test A {test_A} -> {descending_order(test_A)} : should be 54421")
print(f"Test B {test_B} -> {descending_order(test_B)} : should be 654321")
print(f"Test C {test_C} -> {descending_order(test_C)} : should be 987654321")
```

an equals method, to check that two vectors that have the same components are equal  

Note: the test cases will utilize the user-provided equals method.

```python
from math import sqrt 

class Vector:
    def __init__(self, the_list):
        self.list = the_list

    def __is_equal_length(self, ohter_vector):
        ## Checks if the lists are equal in length
        if len(self.list) != len(ohter_vector.list):
            raise ValueError("Lists are different sizes")
        
    def __is_Vector_type(self, ohter_vector):
        if not isinstance(ohter_vector, Vector):
            raise TypeError("Expects a Vector object")
    
    def add(self, ohter_vector):
        self.__is_Vector_type(ohter_vector)
        self.__is_equal_length(ohter_vector)

        return Vector([self.list[i] + ohter_vector.list[i] for i in range(len(self.list))])
    
    def subtract(self, ohter_vector):
        self.__is_Vector_type(ohter_vector)
        self.__is_equal_length(ohter_vector)

        return Vector([self.list[i] - ohter_vector.list[i] for i in range(len(self.list))])
    
    def dot(self, ohter_vector):
        self.__is_Vector_type(ohter_vector)
        self.__is_equal_length(ohter_vector)

        return sum([self.list[i] * ohter_vector.list[i] for i in range(len(self.list))])
    
    def norm(self):
        return sqrt(sum(x * x for x in self.list))
    
    def equals(self, ohter_vector):
        return isinstance(ohter_vector, Vector) and self.list == ohter_vector.list
    
    def __str__(self):
        return f"({','.join(str(x) for x in self.list)})"
    

a = Vector([1, 2, 3])  
b = Vector([3, 4, 5])  
c = Vector([5, 6, 7, 8])  

test = a.add(b)      # should return a new Vector([4, 6, 8])  
print(test.list)
test = a.subtract(b) # should return a new Vector([-2, -2, -2])
print(test.list)
test =  a.dot(b)      # should return 1*3 + 2*4 + 3*5 = 26
print(test)  
test = a.norm()      # should return sqrt(1^2 + 2^2 + 3^2) = sqrt(14)
print(test)  
#a.add(c)      # raises an exception  
```
